```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 5, 10, -30
```

```python
def func(x):
    return a * x ** 2 + b * x + c
```

```python
limit = 5
step = 0.0001
```

```python
x = np.arange(-limit, limit, step)
```

```python
def take_roots(a, b, c) -> tuple:
    discr = (b ** 2 - 4 * a * c)
    if discr > 0:
        x1 = (-b + discr ** 0.5) / (2 * a)
        x2 = (-b - discr ** 0.5) / (2 * a)
        return x1, x2
    elif discr == 0:
        x = -b / (2 * a)
        return (x,)
    else:
        return (None,)
```

```python
roots = take_roots(a, b, c)
```

```python
min_y = min(func(x))
min_x = take_roots(a, b, c - min_y)[0]
```

```python
print(min_x, min_y)
```

```python
x_down_pos = np.arange(-limit, min(roots) + step, step)
x_down_neg = np.arange(min(roots), min_x + step, step)
x_up_neg = np.arange(min_x, max(roots) + step, step)
x_up_pos = np.arange(max(roots), limit + step, step)
```

```python
plt.rcParams['lines.linestyle'] = '-'
plt.plot(x_down_pos, func(x_down_pos), 'b', label = '  ')
plt.plot(x_up_pos, func(x_up_pos), 'r', label = '  ')
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(x_down_neg, func(x_down_neg), 'b', label = '  ')
plt.plot(x_up_neg, func(x_up_neg), 'r', label = '  ')

if len(roots) == 2:
    plt.plot(roots[0], 0, 'yo', label = f' ({roots[0]},{roots[1]})')
    plt.plot(roots[1], 0, 'yo')
else:
    if roots[0] != None:
        plt.plot(roots[0], 0, 'yo', label = f' ({roots[0]},{roots[1]})')
        
plt.plot(min_x, min_y, 'gx', label = f'  ({min_x},{min_y})')
plt.legend()
plt.grid()
plt.show()
```
