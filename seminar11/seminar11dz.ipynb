```python
import math
import matplotlib.pyplot as plt

def f(x):
    return -12 * (x ** 4) * math.sin(math.cos(x)) - 18 * (x ** 3) + 5 * x * x + 10 * x - 30

n = 10000
xmin = -15
xmax = 15
for i in range(n):
    x0 = xmin + ( i      / n) * (xmax - xmin)
    x1 = xmin + ((i + 1) / n) * (xmax - xmin)
    f0 = f(x0)
    f1 = f(x1)
    plt.plot(x0, f(x0), 'r.')
    if(f0 * f1 < 0):
        print(f'Корень = {x0}')

plt.plot(0, f(0), 'gx', label = f'Эктримум функции ({0},{f(0)})')
plt.legend()
plt.grid()
plt.show()
```
